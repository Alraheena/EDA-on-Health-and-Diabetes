# Health & Lifestyle Data for Diabetes 

##### How Lifestyle, Demographics, and Clinical Health Indicators Contribute to Diabetes Risk

### overview

This dataset captures detailed health, lifestyle, and clinical information from a large population of individuals, designed to analyze and predict diabetes risk. Each record represents a comprehensive profile of a patient, including demographic attributes, lifestyle behaviors, family medical background, physiological measurements, and clinical test results.

The dataset includes variables such as age, gender, ethnicity, education level, income, employment status, smoking and alcohol habits, diet and physical activity, sleep patterns, body mass index (BMI), blood pressure, cholesterol levels, glucose, insulin, HbA1c, and family or personal medical history. Most importantly, it contains target variables — diagnosed_diabetes (binary: 0 = No, 1 = Yes) and diabetes_stage (categorical: No Diabetes, Pre-Diabetes, Type 1, Type 2, Gestational) — indicating the presence and stage of diabetes.

The objective of analyzing this dataset is to uncover patterns and risk factors associated with diabetes. By examining correlations between demographic characteristics, lifestyle choices, and clinical measurements, this analysis aims to provide insights into high-risk populations, lifestyle interventions, and potential predictive indicators to support better health management and disease prevention strategies.

### Column Description

Age---->
Age of the individual in years.


gender--->
Biological sex of the individual (Male/Female/Other).


ethnicity--->
Ethnic background of the individual.


education_level--->
Highest education attained (No formal, High school, Graduate, etc.).


income_level--->
Economic status categorized into income groups (Low, Medium, High).


employment_status--->
Employment type or occupational status (Employed, Unemployed, Retired, Student).


smoking_status--->
Smoking behavior (Never, Former, Current).


alcohol_consumption_per_week--->
Average alcoholic drinks consumed per week.


physical_activity_minutes_per_week--->
Minutes spent in physical activity weekly.


diet_score--->
Numerical score representing dietary quality (higher = healthier).


sleep_hours_per_day--->
Average number of sleeping hours per day.


screen_time_hours_per_day--->
Average hours of daily screen usage.

bmi--->
Body Mass Index (BMI) measured in kg/m².


waist_to_hip_ratio--->
Ratio of waist circumference to hip circumference.


heart_rate--->
Resting heart rate (beats per minute).


cholesterol_total--->
Total cholesterol level (mg/dL).


hdl_cholesterol--->
High-density lipoprotein (HDL) cholesterol (mg/dL).


ldl_cholesterol--->
Low-density lipoprotein (LDL) cholesterol (mg/dL).


triglycerides--->
Triglyceride level in the blood (mg/dL).


glucose_fasting--->
Fasting blood glucose level (mg/dL).


glucose_postprandial--->
Post-meal glucose level (mg/dL).


insulin_level--->
Blood insulin concentration (µU/mL).


hba1c--->
Glycated hemoglobin percentage (HbA1c %).


diabetes_risk_score--->
Composite risk score estimating diabetes likelihood.


diabetes_stage--->
Indicates disease stage (No Diabetes, Pre-Diabetes, Type 1, Type 2, Gestational).


diagnosed_diabetes-->
Target variable showing diabetes diagnosis (0 = No, 1 = Yes)

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns


for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

## Conclusions & Recommendations

**1. Age is the Strongest Non-Modifiable Risk Factor**

Observation: Diabetes prevalence doubles after age 45 and peaks in 60+ population

Recommendation:
- Mandatory annual screening for all individuals over 45 years
- Develop age-specific prevention programs targeting middle-aged adults

**2. Central Obesity More Important Than Overall Weight**

Observation: Waist-to-hip ratio predicts diabetes risk better than BMI alone

Recommendation:
- Include waist circumference measurements in routine health check-ups
- Target abdominal fat reduction specifically in weight management programs

**3. Physical Inactivity is the Most Critical Modifiable Factor**

Observation: Sedentary individuals have 40% higher diabetes rates than active counterparts

Recommendation:
- Implement workplace activity programs and community exercise initiatives
- Prescribe physical activity as medical treatment for pre-diabetic patients

**4. Early Detection Through HbA1c Monitoring**

Observation: HbA1c shows strongest correlation with diabetes diagnosis among all clinical markers

Recommendation:
- Routine HbA1c testing for high-risk individuals (BMI >25, age >45)
- Establish clear intervention thresholds at pre-diabetes stage (HbA1c 5.7-6.4%)

**5. Multiple Risk Factors Have Multiplicative Effect**

Observation: Combinations of poor diet, inactivity and smoking increase diabetes risk exponentially

Recommendation:
- Implement integrated lifestyle programs addressing multiple factors simultaneously
- Use risk scoring systems to identify highest-priority individuals


**6. Ethnic Disparities Require Targeted Approaches**

Observation: Certain ethnic groups show significantly higher diabetes prevalence

Recommendation:
- Develop culturally tailored education and prevention programs
- Train healthcare providers on ethnicity-specific risk factors

**7. Diet Quality Matters More Than Quantity**

Observation: Diet score correlates more strongly with diabetes risk than total calorie intake

Recommendation:
- Focus nutrition education on food quality rather than just calorie counting
- Promote whole foods and reduce processed food consumption

**8. Systematic Screening Prevents Late Diagnosis**

Observation: Many cases are detected only after complications appear

Recommendation:
- Implement systematic risk-based screening in primary care settings
- Use electronic health records to flag high-risk patients automatically
